```python
!pip install -q kokoro>=0.9.2 soundfile
!apt-get -qq -y install espeak-ng > /dev/null 2>&1
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch
```

```python
# lang_code 選 'a' 代表英文
pipeline = KPipeline(lang_code='a')

VOICE_MAP = {
    "A": "af_heart",    # 主持人
    "B": "af_bella",    # 來賓
}

dialogue = [
    ("A", "Welcome to today’s podcast! We’re going to talk about Kokoro TTS."),
    ("B", "Thanks for having me. I’ve been really impressed by how lightweight the model is."),
    ("A", "Right? Only 82 million parameters, but the quality rivals much larger models."),
    ("B", "Absolutely. And the Apache license means it can be used basically anywhere."),
]

for i, (speaker, text) in enumerate(dialogue):
    print(f"[{i}] {speaker}: {text}")

    generator = pipeline(text, voice=VOICE_MAP[speaker])

    for _, _, audio in generator:
        display(Audio(data=audio, rate=24000, autoplay=(i == 0)))

        outfile = f"{i:02d}_{speaker}.wav"
        sf.write(outfile, audio, 24000)
        print(f"Saved: {outfile}")
```

```python
from pydub import AudioSegment
import os

def merge_wavs(wav_files, output="final.wav"):
    audio = AudioSegment.empty()

    for wav in wav_files:
        print("Adding:", wav)
        segment = AudioSegment.from_wav(wav)
        audio += segment

    audio.export(output, format="wav")
    print("Saved merged file:", output)

# 例如：合併你的對話 output
wav_files = sorted([f for f in os.listdir(".") if f.endswith(".wav")])
merge_wavs(wav_files, "podcast_dialogue.wav")
```
